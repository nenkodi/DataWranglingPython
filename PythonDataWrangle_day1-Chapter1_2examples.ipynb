## Dictionary

This where we can add additional information to an element in a list. It is base don KEY/VALUE pairs:

```python
menagerie = ['lion', 'giraffe', 'elephant']
```

```python
menagerie = {'lion': 'fierce', 'giraffe':'tall', 'elephant':'big'}
```

```python
menagerie['lion']
```

```python
menagerie.get('lion')
```

The above example is showing two possible calls.

```python
menagerie['monkey']='an animal that hangs on trees' #example on how to add an element to the dictionary
```

```python
menagerie
```

```python
del menagerie['elephant']
```

```python
for animal in menagerie:
    print(animal)
```

```python
for animal in menagerie:
    print(animal,menagerie[animal]) #this prints the key and the value
```
